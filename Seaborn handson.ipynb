# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('Preplacement.csv')
```

```python
dataset
```

```python
# Number of rows and Columns
# we can use dataset.shape also
print("The Number of Row:",dataset.shape[0])
print("The Number of Column",dataset.shape[1])
```

# Univarite Analysis

```python
import seaborn as sb
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
# to see how many time frequency got repeated
```

# Bivarite Analysis : Distribution Plot

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset)
plt.show()
# to see the histogram and scatter plot
```

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='hex')
plt.show()
# to see in hexa and based on hex color dendity then there is repetation
```

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='kde')
plt.show()
# to see the Kernel density view to know the maximum repetation based on curve
```

```python
#Pair plot
sb.pairplot(dataset,hue = 'gender',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
# hue ==differentiate color
# diag_kind =kernel or hue
# to view in different view
```

# For Categorical Plot

```python
sb.stripplot(x = "ssc_b", y = "ssc_p", data = dataset)
plt.show()
# for the same group
```

```python
sb.stripplot(x = "gender", y = "ssc_p", data = dataset,jitter=True)
plt.show()
# to add random noise (jitter) to data points. This helps prevent overlapping points in scatter plots
```

```python
sb.swarmplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
# it will avoid the overlap so making it easier to see individual data points
```

# Distribution Observation

```python
sb.violinplot(x = "gender", y = "ssc_p", data = dataset)
plt.show()
# x= categorical and Y= quantitative to see the density
```

```python
sb.violinplot(x = "specialisation", y = "salary", data = dataset,hue="gender")
plt.show()
# same as 14 and additionally we giving gender to segeregate the data, know more depth of the given condition 
```

```python
sb.boxplot(x = "ssc_b", y = "ssc_p", data = dataset,hue="gender" )
plt.show()
# percentile concept from start to end in y axis 1st line is Q1, 2nd line is Q2, 3rd line is Q3, 4th line Q4 and dot represents outliers
```

```python
# Bar Plot
sb.barplot(x = "ssc_b", y = "ssc_p", data = dataset,hue='gender')
plt.show()
# based on data it shows in bars
```

```python
sb.countplot(x = "ssc_b",data = dataset,hue='gender');
plt.show()
#it gives the count of the data
```

```python
sb.pointplot(x = "ssc_p",y="hsc_b",data =dataset,hue='gender')
plt.show()
# to show the average of numerical variation for each category 
```

```python
sb.boxplot(data = dataset, orient = "h")
plt.show()
# same as box plot but represent in horizontal view
```
